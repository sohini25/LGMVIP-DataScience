```python
#shape of dataset
X= data.drop(['outputs'],axis=1)
y=data['outputs']
print(X.shape)
print(y.shape)
```

```python
#Splitting Dataset into test and train

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4 )
```

```python
#training set
print(X_train.shape)
print(y_train.shape)
#testing set
print(X_test.shape)
print(y_test.shape)
```

```python
#Selecting different no of k in range [1,26] and choose the one with high accuracy

k_range = list(range(1,26))
scores = []
for i in k_range:
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))

    plt.title('Accurate scores for k values of knn')
plt.xlabel('Number of Neighbors K')
plt.ylabel('Accurate scores')
plt.plot(k_range, scores)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score

# Instantiate learning model (k = 13)
classifier = KNeighborsClassifier(n_neighbors=3)

# Fitting the model
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
```
